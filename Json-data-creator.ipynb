```python
import json
import os
def append_data_to_file(new_data, filename):
    try:
        with open(filename, 'r') as file:
            existing_data = json.load(file)
    except FileNotFoundError:
        existing_data = []

    existing_data.append(new_data)

    with open(filename, 'w') as file:
        json.dump(existing_data, file, indent=4)
```

```python
def ret_qas():
    qas=[]
    def ret_qas_more(answers):
        print('Answer wont be changed for paraphased qn')
        data_={ 
            "id":'dummy_id',  
            "is_impossible": False,  
            "question": input("Enter Question: "),  
            "answers":answers
        }
        return data_
    def ret_ans():
        nested=[]
        more_ans='yes'
        while (more_ans =='yes'):
            answer=input("Enter Answer: ")
            instant={ 
                    "text": answer,  
                    "answer_start": int( give_index(answer)),
                } 
            nested.append(instant)
            more_ans=input("More answer?")
        return nested
    rem='yes'
    while rem !='no':
        data_={ 
            "id":'dummy_id',  
            "is_impossible": False,  
            "question": input("Enter Question: "),  
            "answers":ret_ans()
        }
        qas.append(data_)
        gate=input('Paraphased question remaining')
        while gate=='y':
            qas.append(ret_qas_more(data_['answers']))
            gate=input('Paraphased question remaining')
    
        rem=input('Remaining?')
    return qas
    
        
```
